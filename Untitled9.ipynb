```python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048 # 문맥 길이
dtype = None # 자동으로 하드웨어에 맞춤
load_in_4bit = True # 메모리 절약을 위한 4비트 로딩

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gemma-2-9b-it-bnb-4bit", # Gemma 2 9B 버전 추천
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

# 파인튜닝 설정 (LoRA 적용)
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # 숫자가 높을수록 더 정교하게 배우지만 메모리를 많이 씀
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth", # 메모리 최적화
)
```

```python
# Gemma의 대화 양식(Chat Template) 적용
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma", # Gemma 형식 사용
)

def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
    return { "text" : texts, }

# 예시 데이터셋 로드 (본인의 데이터셋으로 교체 가능)
from datasets import load_dataset
dataset = load_dataset("philschmid/gemma-2-27b-it-sharegpt", split = "train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60, # 테스트를 위해 짧게 설정 (실제론 300~1000 권장)
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps = 1,
        output_dir = "outputs",
    ),
)

trainer.train()
```

```python
# 1. 추론 모드 활성화
FastLanguageModel.for_inference(model)

# 2. 질문 던지기
messages = [
    {"role": "user", "content": "너는 이제부터 나의 파인튜닝된 비서야. 인사해줘!"},
]
inputs = tokenizer.apply_chat_template(messages, add_generation_prompt = True, return_tensors = "pt").to("cuda")

outputs = model.generate(input_ids = inputs, max_new_tokens = 64)
print(tokenizer.batch_decode(outputs))

# 3. 구글 드라이브에 저장 (나중에 또 쓰기 위해)
model.save_pretrained_merged("/content/drive/MyDrive/gemma_model", tokenizer, save_method = "merged_16bit")
```
